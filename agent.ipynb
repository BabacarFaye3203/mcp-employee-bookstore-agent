```python
from langchain.agents import create_agent
from langchain_mcp_adapters.client import MultiServerMCPClient
from dotenv import load_dotenv
from langchain_ollama import ChatOllama
from  langchain.messages import SystemMessage, HumanMessage, AIMessage
from IPython.display import display, Markdown
import asyncio
```

```python
llm=ChatOllama(model="llama3.2:latest", temperature=0.9)
Res=llm.invoke([SystemMessage(content="You are a helpful assistant that answers questions about the world."), HumanMessage(content="What is the capital of France?")])
print(display(Markdown(f"**Test Response:** {Res.content}")))
```

```python
mcp_client=MultiServerMCPClient({
    "mcp_server":{
        "transport":"streamable-http",
        "url":"http://localhost:2400/mcp"

    }
})
```

```python
tools=await mcp_client.get_tools()
```

```python
tools
```

```python
agent=create_agent(model=llm, tools=tools, system_prompt="You are a helpful assistant.your role is to assist users with their queries by providing accurate and helpful information in json format using the available tools.")
```

```python
result=await agent.ainvoke({
    "messages": [HumanMessage(content="donne l'employé qui a l'id 1")],
    "stream": True
})
print(display(Markdown(f"**Response:** {result['messages'][-1].content}")))
```
